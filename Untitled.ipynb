```python
# qn 2 cubic spline

import numpy as np
import matplotlib.pyplot as plt
from cubicSpline import *

def evalSpline(xData,yData,k,x):
    
    def findSegment(xData,x):
        iLeft = 0
        iRight = len(xData)- 1
        while 1:
            if (iRight-iLeft) <= 1: return iLeft
            i = int((iLeft + iRight)/2)
            if x < xData[i]:
                iRight = i
            else: 
                iLeft = i
    
    i = findSegment(xData,x)
    h = xData[i] - xData[i+1]
    y = ((x - xData[i+1])**3/h - (x - xData[i+1])*h)*k[i]/6.0 \
      - ((x - xData[i])**3/h - (x - xData[i])*h)*k[i+1]/6.0   \
      + (yData[i]*(x - xData[i+1])                            \
       - yData[i+1]*(x - xData[i]))/h
    return y

f = open("2.txt", "r")
nPoints = int(f.readline())
tData = []
xData = []
yData = []

# storing all x and y values of the particle with t as x axis
tDiff = 2.0 / (nPoints - 1)
for i in range(0, nPoints):
    tData.append(tDiff * i)
    arrPtData = f.readline().split()
    xData.append(arrPtData[0])
    yData.append(arrPtData[1])

# data pt at specific t
ActualData = f.readline().split()
tActual = float(ActualData[0])
xActual = float(ActualData[1])
yActual = float(ActualData[2])
    
tData = np.array(tData, float)
xData = np.array(xData, float)
yData = np.array(yData, float)

kx = curvatures(tData,xData)
ky = curvatures(tData,yData)

XPredict = evalSpline(tData, xData, kx, tActual)
print("Actual X value: ", xActual)
print("Predicted X value: ", XPredict)

YPredict = evalSpline(tData, yData, ky, tActual)
print("Actual Y value: ", yActual)
print("Predicted Y value: ", YPredict)

input("Press Enter to continue...")
```

```python
# qn3 Chebyshev polynomials
from math import cos, acos
from scipy.interpolate import lagrange

f = open("4.txt", "r")
nPoints = int(f.readline())
tData = []
xData = []
yData = []

# storing all x and y values of the particle with t as x axis
tDiff = 2.0 / (nPoints - 1)
for i in range(0, nPoints):
    tData.append(tDiff * i)
    arrPtData = f.readline().split()
    xData.append(float(arrPtData[0]))
    yData.append(float(arrPtData[1]))

# data pt at specific t
ActualData = f.readline().split()
tActual = float(ActualData[0])
xActual = float(ActualData[1])
yActual = float(ActualData[2])
    
tData = np.array(tData, float)
xData = np.array(xData, float)
yData = np.array(yData, float)

xcoef = np.polynomial.chebyshev.chebfit(tData, xData, nPoints - 1)
xPredict = np.polynomial.chebyshev.chebval(tActual, xcoef)
print("Predicted X value: ", xPredict)
print("Actual X value: ", xActual)

ycoef = np.polynomial.chebyshev.chebfit(tData, yData, nPoints - 1)
yPredict = np.polynomial.chebyshev.chebval(tActual, ycoef)
print("Predicted Y value: ", yPredict)
print("Actual Y value: ", yActual)

input("Press Enter to continue...")
```
